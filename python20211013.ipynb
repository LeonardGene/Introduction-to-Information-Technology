# **IF**

```python
a = 8
b = 27
if a < b:
  print("a is less than b")
```

# **ELIF**

```python
a = 8
b = 27
if b < a:
  print("b is greater than a")
elif a < b:
  print("a is less than b")
```

# **ELSE**

```python
a = 2002
b = 2002
if b < a:
  print("b is greater than a")
elif a < b:
  print("a is less than b")
else:
  print("a and b are equal")
```

# **LEAP YEAR**
#### https://en.wikipedia.org/wiki/Leap_year

```python
year = eval(input("Please input the year: "))

if ((year%400==0) or (year%4==0 and year%100!=0)):
  print("{0} is a leap year.".format(year))
else:
  print("{0} is not a leapyear.".format(year))
```

# **FOR LOOPS**
#### https://www.w3schools.com/python/python_for_loops.asp


# Range


```python
for x in range(28):
  print(x)
```

# Break Statement

```python
colors = ["red", "blue", "yellow"]
for x in colors:
  print(x)
  if x == "blue":
    break
```

# Continue Statement

```python
colors = ["red", "blue", "yellow", "violet", "green"]
for x in colors:
  if x == "yellow":
    continue
  print(x)
```

# While Loops

```python
a = 8
while a < 27:
  print(a)
  a += 2
```

```python
a = 8
while a < 27:
  a += 2
  if a == 16:
    continue
  print(a)
```

```python
a = 27
while a > 8:
  print(a)
  if a == 20:
    break
  a -= 2
```

# Infinite Loop

```python
a = 8
while a < 27:
  print(a)
  if a == 5:
    continue
#   break
    a += 3
```
